**Step 2: Wrap your method chain(s) in a function**

```python
def load_and_process(address):
    df1 = (
          pd.read_csv(address)
          .dropna(subset=['bmi'])
          .dropna(subset=['sex'])
          .round({"bmi":1})
          .sort_values('bmi',ascending=True)
          .reset_index(drop=True)
    )
    return df1
```
